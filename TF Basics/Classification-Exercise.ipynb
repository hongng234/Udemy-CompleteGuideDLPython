### THE DATA

** Read in the census_data.csv data with pandas**

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline

from keras import backend as K

#Clear TF memory
cfg = K.tf.ConfigProto()
cfg.gpu_options.allow_growth = True
K.set_session(K.tf.Session(config=cfg))
```

```python
df = pd.read_csv('census_data.csv')
```

```python
df['income_bracket']
```

** TensorFlow won't be able to understand strings as labels, you'll need to use pandas .apply() method to apply a custom function that converts them to 0s and 1s. This might be hard if you aren't very familiar with pandas, so feel free to take a peek at the solutions for this part.**

** Convert the Label column to 0s and 1s instead of strings.**

```python
df['income_bracket'].unique()
```

```python
def label_fix(label):
    if label == ' <=50K':
        return 0
    else:
        return 1
```

```python
df['income_bracket'] = df['income_bracket'].apply(label_fix)
```

```python
df
```

### Perform a Train Test Split on the Data

```python
x_data = df.drop('income_bracket', axis=1)
y_data = df['income_bracket']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.30, random_state=42)
```

### Create the Feature Columns for tf.esitmator

** Take note of categorical vs continuous values! **

```python
df.columns
```

** Create the tf.feature_columns for the categorical values. Use vocabulary lists or just use hash buckets. **

```python
gender = tf.feature_column.categorical_column_with_vocabulary_list('gender', vocabulary_list=['Male','Female'])
```

```python
workclass = tf.feature_column.categorical_column_with_hash_bucket('workclass', hash_bucket_size=1000)
education = tf.feature_column.categorical_column_with_hash_bucket('education', hash_bucket_size=1000)
marital_status = tf.feature_column.categorical_column_with_hash_bucket('marital_status', hash_bucket_size=1000)
occupation = tf.feature_column.categorical_column_with_hash_bucket('occupation', hash_bucket_size=1000)
relationship = tf.feature_column.categorical_column_with_hash_bucket('relationship', hash_bucket_size=1000)
race = tf.feature_column.categorical_column_with_hash_bucket('race', hash_bucket_size=1000)
native_country = tf.feature_column.categorical_column_with_hash_bucket('native_country', hash_bucket_size=1000)
```

** Create the continuous feature_columns for the continuous values using numeric_column **

```python
age = tf.feature_column.numeric_column('age')
education_num = tf.feature_column.numeric_column('education_num')
capital_gain = tf.feature_column.numeric_column('capital_gain')
capital_loss = tf.feature_column.numeric_column('capital_loss')
hours_per_week = tf.feature_column.numeric_column('hours_per_week')
```

** If using the DNNClassifier => embedding_columns **

```python
len(df['workclass'].unique())
```

```python
#If using DNNClassifier
embedding_gender = tf.feature_column.embedding_column(categorical_column=gender, dimension=2)
embedding_workclass = tf.feature_column.embedding_column(categorical_column=workclass, dimension=9)
embedding_education = tf.feature_column.embedding_column(categorical_column=education, dimension=16)
embedding_marital_status = tf.feature_column.embedding_column(categorical_column=marital_status, dimension=7)
embedding_occupation = tf.feature_column.embedding_column(categorical_column=occupation, dimension=15)
embedding_relationship = tf.feature_column.embedding_column(categorical_column=relationship, dimension=6)
embedding_race = tf.feature_column.embedding_column(categorical_column=race, dimension=5)
embedding_native_country = tf.feature_column.embedding_column(categorical_column=native_country, dimension=42)
```

** Put all these variables into a single list with the variable name feat_cols **

```python
feat_cols = [age, workclass, education, education_num, marital_status,
             occupation, relationship, race, gender, capital_gain,
             capital_loss, hours_per_week, native_country]

embedding_feat_cols = [age, embedding_workclass, embedding_education, education_num, embedding_marital_status,
                       embedding_occupation, embedding_relationship, embedding_race, embedding_gender, capital_gain,
                       capital_loss, hours_per_week, embedding_native_country]
```

```python
feat_cols
```

```python
embedding_feat_cols
```

### Create Input Function

** Batch_size is up to you. But do make sure to shuffle!**

```python
input_func = tf.estimator.inputs.pandas_input_fn(x_train, y_train, batch_size=100, num_epochs=None, shuffle=True)
```

#### Create your model with tf.estimator

**Create a LinearClassifier.(If you want to use a DNNClassifier, keep in mind you'll need to create embedded columns out of the cateogrical feature that use strings, check out the previous lecture on this for more info.)**

```python
# model = tf.estimator.LinearClassifier(feature_columns=feat_cols)
model = tf.estimator.DNNClassifier(hidden_units=[10,10,10], feature_columns=embedding_feat_cols)
```

** Train your model on the data, for at least 5000 steps. **

```python
model.train(input_fn=input_func, steps=5000)
```

### Evaluation

** Create a prediction input function. Remember to only supprt X_test data and keep shuffle=False. **

```python
pred_input_function = tf.estimator.inputs.pandas_input_fn(x_test, batch_size=len(x_test), shuffle=False, num_epochs=1)
```

** Use model.predict() and pass in your input function. This will produce a generator of predictions, which you can then transform into a list, with list() **

```python
preds = list(model.predict(input_fn=pred_input_function))
```

** Each item in your list will look like this: **

```python
preds
```

** Create a list of only the class_ids key values from the prediction list of dictionaries, these are the predictions you will use to compare against the real y_test values. **

```python
final_preds = [pred['class_ids'][0] for pred in preds]
```

```python
final_preds
```

** Import classification_report from sklearn.metrics and then see if you can figure out how to use it to easily get a full report of your model's performance on the test data. **

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, final_preds))
```
